```python
import os
import cv2
import mediapipe as mp
import time
import numpy as np
import cmath
```

```python
groups = os.listdir('Resources/')[1:]
classes = {}
for index, value in enumerate(groups):
    classes[value] = index

recordedType = groups[0]
twohands = True

recordedType
# classes
```

```python
mpHands = mp.solutions.hands

hands = mpHands.Hands(
    static_image_mode=False,
    max_num_hands=2,
    min_detection_confidence=0.5)


mpFace = mp.solutions.face_detection

face = mpFace.FaceDetection(
    model_selection=1,
    min_detection_confidence=0.5)

mpDraw = mp.solutions.drawing_utils
```

```python
cap = cv2.VideoCapture(0)

pTime = 0
cTime = 0
count = 0
save = True

dataset = []

while True:
    
    success, img = cap.read()
    resolution = img.shape
    
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    handResults = hands.process(imgRGB)
    faceResults = face.process(imgRGB)
    
    if handResults.multi_hand_landmarks and faceResults.detections:
        count += 1
        faceCenter = [mpFace.get_key_point(faceResults.detections[0],
                                          mpFace.FaceKeyPoint.NOSE_TIP).x * resolution[1],
                     
                     mpFace.get_key_point(faceResults.detections[0],
                                          mpFace.FaceKeyPoint.NOSE_TIP).y * resolution[0]]
        
        handcords = []
        
        for handLms in handResults.multi_hand_landmarks:
            for id, lm in enumerate(handLms.landmark):
                if id == 0:
                    base = [lm.x*resolution[1]-faceCenter[0], lm.y*resolution[0]-faceCenter[1]]
                else:
                    cord = [lm.x*resolution[0]-faceCenter[0],lm.y*resolution[1]-faceCenter[1], lm.z * 100]
                    cv2.circle(img, (int(cord[0]), int(cord[1])), 10, (0,0,0), 5)
                    cv2.circle(img, (int(base[0]), int(base[1])), 10, (255,0,0), 5)
                    
                    handcords += cord
                    
            mpDraw.draw_landmarks(img, handLms, mpHands.HAND_CONNECTIONS)
        
        if len(handcords) not in [120, 60]:
            print("error has occurred:", len(handcords))
            cv2.imshow("img", img)
            continue
        
        if len(handcords) == 60 and twohands:
            print("only one hand detected")
            cv2.imshow("img", img)
            continue
        
        
        if len(handcords) == 120 and np.random.randint(2) == 1:
            tmp = handcords[:60]
            handcords[:60] = handcords[60:] 
            handcords[60:] = tmp
            
        if count%7 == 0 and save:
            y = np.eye(1)
            dataset.append([ np.array(handcords), y] )
            
        cv2.circle(img, (300,300), 10, (0, 100,100), 5)
                
    cTime= time.time()
    fps = 1/(cTime-pTime)
    pTime = cTime
    
    cv2.putText(img, str(int(fps)) + f" {count}", (10,70), cv2.FONT_HERSHEY_PLAIN, 3, (255,255,255), 3)
    
    cv2.imshow("img", img)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release() 
```

```python
print(np.array(dataset).shape, recordedType)
if save:
    np.save(f"Resources/{recordedType}/{recordedType}.npy", dataset)
```
